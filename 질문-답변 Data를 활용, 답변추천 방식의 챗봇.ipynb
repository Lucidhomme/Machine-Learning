```python
### 질문DF 답변DF 구분
q_df = fsh[fsh['QA여부'] == 'q']
a_df = fsh[fsh['QA여부'] == 'a']
```

```python
## 특정 컬럼 추출
qqq = q_df[['발화문']]
aaa = a_df[['발화문']]
```

```python
### 각 컬럼명 변경
qqq.columns = ['Q']
aaa.columns = ['A']
qqq
```

```python
### Q칼럼과 A칼럼 df 병합
QA = pd.concat([qqq, aaa], axis = 1)
QA
```

```python
### 텍스트 유사도를 별도 컬럼에 저장
sTe = Pororo(task="sentence_embedding", lang="ko")
QA['EmbVector'] = QA['Q'].progress_map(lambda x : sTe(x))
```

```python
QA.head()
```

```python
### EmbVector를 Tensor로 변환
EmbData = torch.tensor(QA['EmbVector'].tolist())
```

```python
### 코사인유사도로 계산
def chat(sent="0"):
  while 1:
    q = input("user > ").strip()
    if q == "quit":
      break
    a = ""
    # Pororo Sentense Embedding으로 텍스트 유사도를 구합니다.
    q = sTe(q)
    # 질문을 Tensor로 바꿉니다.
    q = torch.tensor(q)
    # 코사인 유사도 
    cos_sim = util.pytorch_cos_sim(q, EmbData)

    #유사도가 가장 비슷한 질문 인덱스를 구합니다.
    best_sim_idx = int(np.argmax(cos_sim))

    # 질문의 유사도와 가장 비슷한  답변 제공
    answer = QA['A'][best_sim_idx]
    print(answer)
```

```python
chat()
```
