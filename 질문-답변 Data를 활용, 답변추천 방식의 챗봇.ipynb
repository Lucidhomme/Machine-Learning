```python
### ì§ˆë¬¸DF ë‹µë³€DF êµ¬ë¶„
q_df = fsh[fsh['QAì—¬ë¶€'] == 'q']
a_df = fsh[fsh['QAì—¬ë¶€'] == 'a']
```

```python
## íŠ¹ì • ì»¬ëŸ¼ ì¶”ì¶œ
qqq = q_df[['ë°œí™”ë¬¸']]
aaa = a_df[['ë°œí™”ë¬¸']]
```

```python
### ê° ì»¬ëŸ¼ëª… ë³€ê²½
qqq.columns = ['Q']
aaa.columns = ['A']
qqq
```

```python
### Qì¹¼ëŸ¼ê³¼ Aì¹¼ëŸ¼ df ë³‘í•©
QA = pd.concat([qqq, aaa], axis = 1)
QA
```

```python
### í…ìŠ¤íŠ¸ ìœ ì‚¬ë„ë¥¼ ë³„ë„ ì»¬ëŸ¼ì— ì €ì¥
sTe = Pororo(task="sentence_embedding", lang="ko")
QA['EmbVector'] = QA['Q'].progress_map(lambda x : sTe(x))
```

```python
QA.head()
```

```python
### EmbVectorë¥¼ Tensorë¡œ ë³€í™˜
EmbData = torch.tensor(QA['EmbVector'].tolist())
```

```python
### ì½”ì‚¬ì¸ìœ ì‚¬ë„ë¡œ ê³„ì‚°
def chat(sent="0"):
  while 1:
    q = input("user > ").strip()
    if q == "quit":
      break
    a = ""
    # Pororo Sentense Embeddingìœ¼ë¡œ í…ìŠ¤íŠ¸ ìœ ì‚¬ë„ë¥¼ êµ¬í•©ë‹ˆë‹¤.
    q = sTe(q)
    # ì§ˆë¬¸ì„ Tensorë¡œ ë°”ê¿‰ë‹ˆë‹¤.
    q = torch.tensor(q)
    # ì½”ì‚¬ì¸ ìœ ì‚¬ë„ 
    cos_sim = util.pytorch_cos_sim(q, EmbData)

    #ìœ ì‚¬ë„ê°€ ê°€ì¥ ë¹„ìŠ·í•œ ì§ˆë¬¸ ì¸ë±ìŠ¤ë¥¼ êµ¬í•©ë‹ˆë‹¤.
    best_sim_idx = int(np.argmax(cos_sim))

    # ì§ˆë¬¸ì˜ ìœ ì‚¬ë„ì™€ ê°€ì¥ ë¹„ìŠ·í•œ  ë‹µë³€ ì œê³µ
    answer = QA['A'][best_sim_idx]
    print(answer)
```
